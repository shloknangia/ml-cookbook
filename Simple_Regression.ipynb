```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def predict_sales(radio, weight, bias):
#     return weight*radio + bias
    return np.dot(weight, radio) + bias
```

```python
# MSE = (1/N) sumof(yi - (mxi + b))^2
def cost_function(radio, sales, weight, bias):
    companies = len(radio)
    total_error = 0.0
    for i in range(companies):
        total_error +=  np.power(sales[i] - (np.dot(weight,radio[i]) + bias),2)
    return total_error    
```

```python
def update_weights(radio, sales, weight, bias, learning_rate):
    weight_deriv = 0
    bias_deriv = 0
    companies = len(radio)
    
    for i in range(companies):
        # Calculate partial derivative  
        # -2x(y - ( mx + b ))
        weight_deriv += -2*radio[i]*(sales[i] - (weight*radio[i] + bias))
        
        # -2(y - ( mx + b ))
        bias_deriv += -2*(sales[i] - (weight*radio[i] + bias))
        
    weight -= (weight_deriv/companies) * learning_rate
    bias -= (bias_deriv/companies) * learning_rate
    
    return weight, bias
```

```python
def train(radio, sales, weight, bias, learning_rate, iters):
    cost_history = []
    
    for i in range(iters):
        weight, bias = update_weights(radio, sales, weight, bias, learning_rate)
        
        # calculating cost
        # cost = cost_function(features, targets, weights)
        cost = cost_function(radio, sales, weight, bias)
        cost_history.append(cost)
        
        # Logging
        if i % 10 == 0:
            print("iter: " + str(i) + ", weight: " + str(weight) + ", bias: " + str(bias) + ", cost: " + str(cost))
            plt.plot(radio, sales, 'b.')
            plt.plot(radio, predict_sales(radio, weight, bias))
            plt.show()
            
    return weight, bias, cost_history
```

```python
data = pd.read_csv('./Data/Advertising.csv')
tv = data['TV']
radio = data['radio']
newspaper = data['newspaper']
sales = data['sales']

plt.scatter(radio, sales)
plt.show()


iters = 100
weight, bais, cost_history = train(radio, sales, 0.03, 0.0014, 0.0001, iters)
print("weight: " + str(weight) + " bais: " + str(bais))
```

```python
plt.plot(range(iters), cost_history)
plt.title('Cost in each iteration')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()
```
